# Ejercico 4: Declarar variables tipo entero, flotante, cadena, booleano e imprimir el tipo de cada una

```python
# Declaramos cada tipo de variable
entero = 17
flotante = 1.25
cadena = "Estoy programando en Python!"
booleano = False

# Imprimimos las variables que declaramos usando el type() en cada una para mostrar su tipo
print("Tipo entero:", type(entero))
print("Tipo flotante:", type(flotante))
print("Tipo cadena:", type(cadena))
print("Tipo booleano:", type(booleano))
```

# Ejercicio 5: Escribir un programa que concatene nombre y apellido almacenados en variables

```python
# Variables de nombre y apellido
nombre = "Billy"
apellido = "Peinado"

# Hacemos uso de la concatenacion para unir dos cadenas de texto
nombre_completo = nombre + " " + apellido

print("Nombre completo:", nombre_completo)
```

# Ejercicio 6: Convertir una variable tipo string con valor numerico a entero sumandole 10

```python
# Declaramos el valor numerico en una variable tipo string
string_numero = "110"

# Convertimos el string a entero
int_numero = int(string_numero)

# Hacemos la suma
resultado = int_numero + 10

print("Resultado:", resultado)
```
