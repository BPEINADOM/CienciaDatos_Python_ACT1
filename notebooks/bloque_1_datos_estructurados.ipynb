# Ejercicio 2: Crear una variable tipo texto que represente un comentario libre de un Usuario y contar cuantas palabras tiene

```python
comentario = "Python es bastante interesante"

# split() dividira el texto en palabras
palabras = comentario.split()

# len devuelve el numero de palabras en un solo contenedor, osea el numero de palabras
cantidad = len(palabras)

print("Numero de palabras:", cantidad)
```

# Ejercicio 3: Crear un diccionario con informacion estructurada de un estudiante e imprimir cada valor

```python
# Se crea la informacion del estudiante en un dato estructurado
estudiante = {
    "nombre": "Daniel",
    "edad": "18",
    "carrera" : "Ingenieria de Software"
}

# Imprimimos cada valor usando el print para cada uno llamando al estudiante
print("Nombre:", estudiante["nombre"])
print("Edad:", estudiante["edad"])
print("Carrera:", estudiante["carrera"])
```
