### Ejercicio 7:Calcula el promedio de tres números usando operadores aritméticos

```python
a = 10
b = 20
c = 30

promedio = (a + b + c) / 3
print("El promedio es:", promedio)
```

### Ejercicio 8: Usa operadores de comparación para verificar si un número es mayor que 50.
